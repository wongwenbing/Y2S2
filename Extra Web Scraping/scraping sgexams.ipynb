```python
import requests
import pandas as pd 
import time
import string

from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.chrome.options import Options
```

```python
options=Options()
options.add_argument('--headless=new')
```

```python
driver=webdriver.Chrome()
driver.maximize_window()
time.sleep(3)
url = "https://www.reddit.com/r/technology/top/?t=week"
driver.get(url)
```

```python
letters = list(string.ascii_uppercase)
print(letters)
airlines=[]
```

```python
for letter in letters: 
    letter_link=driver.find_element(By.LINK_TEXT, letter)
    print(letter)
    letter_link.click()
    time.sleep(10)
    main = driver.find_elements(By.CLASS_NAME, "aggregateRow.clearfix")
    for content in main: 
        try: 
            data= {
                "Airline": content.find_element(By.CSS_SELECTOR, ".aggregateColumn_two").text,
                "Score": content.find_element(By.CLASS_NAME, "aggregateColumn_three").text, 
                "Number of Review": content.find_element(By.CLASS_NAME, "aggregateColumn_four").text

            }
        except IndexError:
            continue
        except: 
            continue
        print(data)
        airlines.append(data)
```

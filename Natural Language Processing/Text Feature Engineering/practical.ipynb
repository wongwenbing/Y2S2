Exercise 3c
- use tfidf vectorizer to generate tfidf matrix

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from nltk.tokenize import word_tokenize
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
corpus=['John has some cats.',
'Cat eat fish.',
'I eat a big fish']
```

```python
lemmatizer=WordNetLemmatizer()
stop_words=stopwords.words('English')
newCorpus=[]
for sent in corpus: 
    text=re.sub("[^a-zA-Z0-9]", " ",sent)
    tokens=word_tokenize(text)
    lstToken=[]
    for token in tokens:
        if token not in stop_words:
            token=lemmatizer.lemmatize(token)
        lstToken.append(token)
    newCorpus.append(" ".join(lstToken))
print(newCorpus)
```

```python
tfidf=TfidfVectorizer(lowercase=True, ngram_range=(1,1),
                      stop_words=stopwords.words("english"))
transform=tfidf.fit_transform(newCorpus).toarray()
df=pd.DataFrame(transform,columns=tfidf.get_feature_names_out())
df
```

Task 3: Use regular expressions to remove ‘\n’ character and all words with numbers in the text; save the cleaned lyrics in to a list called corpus

```python
beatles["new_text"]=""
for i in beatles.index: 
    text=re.sub("\n", '', beatles.loc[i,"text"])
    text=re.sub('\w*\d+\w*','',text)
    beatles.loc[i,'new_text'] = text
beatles.head()
corpus=beatles["new_text"].tolist()
corpus
```

Exercise 6
- use pairwise cosine_similarity(), combination(), and CountVectorizer() to compare the text

```python
corpus=['The weather is hot under the sun',
'I make my hot chocolate with milk',
 'One hot encoding',
 'I will have a chai latte with milk',
 'There is a hot sale today']

cv=CountVectorizer(lowercase=True,stop_words='english')
X=cv.fit_transform(corpus).toarray()
dt=pd.DataFrame(X,columns=cv.get_feature_names_out())
```

```python
from itertools import combinations
from sklearn.metrics.pairwise import cosine_similarity

pairs=list(combinations(range(len(corpus)),2))
print(pairs)
cs_list=[]

for (idx_a, idx_b) in pairs: 
    cs=round(cosine_similarity([X[idx_a]],[X[idx_b]])[0][0],5)
    cs_list.append(cs)
    print(f"Cosine similarity for pairs({idx_a},{idx_b}) is {cs}")
```

```python
combos =[(corpus[a_index], corpus[b_index]) for (a_index, b_index) in pairs]
for combo, cs in zip(combos, cs_list):
    print(f"""
String: {combo[0]}
Stirng: {combo[1]}
Cosine Similarity: {cs}\n""")
```

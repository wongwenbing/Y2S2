### Two steps tokenising approach
* Break text into sentence
* Within sentence break into tokens

```python
import nltk 
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize

sentence = """Hi Mr. Smith! I’m going to buy some vegetables (tomatoes and cucumbers) from the store. 
Should i pick up some black-eyed peas as well?"""

setences = sent_tokenize(sentence)

words = []

for s in setences: 
    words.append(word_tokenize(s))
    
print(words)
```

```python
import pandas as pd
```

```python
phone = pd.read_csv("phone_review.csv")
phone
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

phone["Review_ngram"]=""
phone
```

```python
for i in phone.index:
    # print(review)
    words=word_tokenize(phone["Reviews"][i])
    gram=list(ngrams(words,2))
    # print(gram)
    phone.at[i,"Review_ngram"]=gram
```

# Combined

* read news clipping data file, display first 5 rows
* Tokenise each new clipping into sentences and save list of sentences. Name new columns as **sent_news**
* print out sentence token in row 0 and count # of sentences in it
* for each sentence, further tokenise into uni-grams and save into a new_column called word_tokens within same df
* compare first sentence in column **sent_news** and **word_token** in first row
* each sentence, tokenise into **bi-grams**, save into a new column called **bigram_tokens** within same df

```python
news = pd.read_csv("news_clipping.csv")
# news
```

```python
news["sent_news"]=''
```

```python
for i in news.index: 
    sent=sent_tokenize(news["news_snippets"][i])
    # print(sent)
    news.at[i,'sent_news']=sent
news
```

```python
row0=news["sent_news"][0]
for i in range (len(row0)):
    print(f"{i}: {row0[i]}")
```

```python
news["word_tokens"]=''
for i in news.index: 
    row=[]
    for sent in news["sent_news"][i]: 
        # print(sent) 
        words = word_tokenize(sent)
        # gram= list(ngrams(words,1))
        row.append(words)
    print(row)
    news.at[i,'word_tokens']=row

news
    
```

```python
news["bi_grams"]=''
for i in news.index: 
    row=[]
    for sent in news["sent_news"][i]: 
        # print(sent) 
        words = word_tokenize(sent)
        gram= list(ngrams(words,2))
        row.append(gram)
    print(row)
    news.at[i,'bi_grams']=row

news
    
```

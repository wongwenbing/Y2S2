#### (d) - For the Student_Name column, it would be better as two separate columns. Do split each string on the underscore delimiter _ to give a new temporary dataframe called 'temp' with the correct values. Assign the correct column names 'First_Name' and 'Last_Name' to this temporary dataframe

````python
# Type your Q2(d) answer here

# Split Student_Name into First_Name and Last_Name
temp=pd.DataFrame()
names=student["Student_Name"].str.split("_",n=1,expand=True)
temp["First_Name"]=names[0]
temp["Last_Name"]=names[1]
````

#### (e) - Delete the Student_Name from dataframe 'student' and attach/join the temporary dataframe 'temp' from the previous questions in (d) into the dataframe 'student'

````python
# Type your Q2(e) answer here
student.drop("Student_Name",axis=1, inplace=True)
student=pd.concat([student,temp],axis=1)
student
````

#### (f) - Do Standardise all the strings in columns 'First_Name" and "Last_Name" to full uppercase. 'john' will become 'JOHN

````python
# Type your Q2(f) answer here
student["First_Name"]=student["First_Name"].str.upper()
student["Last_Name"]=student["Last_Name"].str.upper() 
student
````

#### (g) - Do compute the age of the students and create it as a new column known as 'Age' in the dataframe.

````python
# Type your Q2(g) answer here

# Convert 'Date_of_Birth' to datetime format
student["Date_of_Birth"]=pd.to_datetime(student["Date_of_Birth"], format='%Y-%m-%d')
student.info()
# Calculate the age and create a new column 'Age'
from datetime import datetime
today=date.today()
student["Age"]=(today.year-student["Date_of_Birth"].dt.year)
student
````

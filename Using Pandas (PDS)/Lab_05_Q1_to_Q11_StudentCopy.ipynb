##### Q11 - Let import dataset from "temperatures.csv" into DataFrame known as temperatures

- a. Display the top 5 and bottom 5 rows of the DataFrame

- e. Setting the index of the DataFrame to be a multi-level index using the columns "country" and "city" and display the Top 5 rows
- f. Sort the DataFrame based on Indexes in _ascending_ order and display the Top 5 rows and display the top 5 rows.
- g. Select rows starting from the combination ("India", "Hyderabad") up to and including the combination ("Iraq", "Baghdad").
- h. Select the rows which filter the rows in the DataFrame based on the specified date range. From 1 Jan 2010 to 31 Jan 2010

```python
# Import your Dataset from "temperatures.csv" into DataFrame known as temperatures
temperatures=pd.read_csv('temperatures.csv')
```

```python
# Type your Q11 (a) answer here
temperatures.head(5)
```

```python
# Type your Q11 (a) answer here
temperatures.tail(5)
```

```python
# Type your Q11 (b) answer here
print(temperatures.shape)
```

```python
# Type your Q11(c) answer here
temperatures=temperatures.drop(axis=1,columns="s/n")
temperatures.head()
```

```python
# Type your Q11(d) answer here
temperatures.describe()
```

```python
temperatures.columns.nlevels
```

```python
# Type your Q11(e) answer here
temperatures_indx=temperatures.set_index(["country","city"])
temperatures_indx.head()
```

```python
# Type your Q11(f) answer here
temperatures_srt=temperatures_indx.sort_index()
temperatures_srt
```

```python
# Type your Q11(g) answer here

temperatures_srt.loc[("India","Hyderabad"):("Iraq","Baghdad")]
```

```python
# Type your Q11(h) answer here
print(temperatures)
temperatures["date"]=pd.to_datetime(temperatures["date"],format="%m/%d/%Y")
temperatures
```

```python
temperatures_bool = temperatures[
    (temperatures["date"] >= "2010-01-01") &
    (temperatures["date"] <= "2010-01-31")
]
temperatures_bool
```
